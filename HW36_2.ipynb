```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

data = pd.read_csv('petrol_consumption.csv')
print(data)
```

```python
X = data.iloc[:, :4]
# print(X)
y = data['Petrol_Consumption']

X_array = np.array(X)
y_array = np.array(y)

model = LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_array, y_array, test_size=0.30, random_state=4)
model.fit(X_train, y_train)
mod_score = model.score(X_test, y_test)
print("Model Score: ", mod_score)
```

```python
# K-Fold Validation
from sklearn.model_selection import KFold, cross_val_score

model1 = LinearRegression()
kfold_validation = KFold(4)
mod_score_k = cross_val_score(model, X_array, y_array, cv=kfold_validation)
print("K-Fold Validation\nCross Validation Scores: ", mod_score_k)
print("Average CV Score: ", mod_score_k.mean())

best_score=0
for i in range(3,10):
    kfold_validation = KFold(n_splits=i)
    mod_score_k = cross_val_score(model1, X_train, y_train, cv=kfold_validation)
    print("Cross Validation Scores: ", mod_score_k)
    print("Average CV Score: ", mod_score_k.mean())
    print("Number of CV Scores used in Average: ", len(mod_score_k))
    if best_score < mod_score_k.mean():
        best_score = mod_score_k.mean()
        best_n_splits = i
print('\nK-Fold Validation\nbest_score =', best_score, '\nbest_n_splits =', best_n_splits)
```

```python
# Shuffle Split CV
from sklearn.model_selection import ShuffleSplit, cross_val_score

model2 = LinearRegression()  

best_score=0
for i in range(2,10):
    shuff_spl_validation = ShuffleSplit(train_size=0.6, test_size=0.3, n_splits=i)
    mod_score_shsp = cross_val_score(model2, X_train, y_train, cv = shuff_spl_validation)
    print("Cross Validation Scores: ", mod_score_shsp)
    print("Average CV Score: ", mod_score_shsp.mean())
    print("Number of CV Scores used in Average: ", len(mod_score_shsp))
    if best_score < mod_score_shsp.mean():
        best_score = mod_score_shsp.mean()
        best_n_splits = i
print('\nShuffle Split CV\nbest_score =', best_score, '\nbest_n_splits =', best_n_splits)
```
